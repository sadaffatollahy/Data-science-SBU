## 3.Is there a significant relationship  between the popularity and artist name.

```python
from scipy.stats import kruskal

ArtistName_groups = []
for ArtistName_type in df['Artist Name(s)'].unique():
    ArtistName_groups.append(df[df['Artist Name(s)'] == ArtistName_type]['Popularity'])

stat, p = kruskal(*ArtistName_groups)
print('Kruskal-Wallis - stat={0:.3f}, p={0:.3f}'.format(stat, p))

if p < 0.05:
    print('There are significant differences in popularity across different artist name .')
else:
    print('There may not be significant differences in popularity across different artist name.')
```

## 6.Is there a significant relationship  between the artist name and artist genres.

```python
# Example of the Chi-Squared Test

from scipy.stats import chi2_contingency

# Create a contingency table with 'artist name' and 'artist genres'
contingency_table = pd.crosstab(df['Artist Name(s)'], df['Artist Genres'])

# Perform the Chi-square test of independence
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p_value)

# Check if there is a significant difference
alpha = 0.05  # Set significance level
if p_value < alpha:
    print("There is a significant difference between 'artist name' and 'artist genres'.")
else:
    print("There is no significant difference between 'artist name' and 'artist genres'.")
```
